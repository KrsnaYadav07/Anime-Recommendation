```python
# Select only the needed columns
new_df = anime_df[[
    "anime_id",
    "anime_url",
    "image_url",
    "title_final",
    "genres",
    "themes",
    "synopsis"
]].copy()
```

```python
new_df = new_df.merge(anime_df[["anime_id", "score", "anime_url", "image_url"]], on="anime_id", how="left")
```

```python
# Drop old duplicated columns if they exist
new_df = new_df.drop(columns=[ "anime_url_x", "image_url_x"], errors=True)

# Rename merged columns to standard names
new_df = new_df.rename(columns={
    "score_y": "score",
    "anime_url_y": "anime_url",
    "image_url_y": "image_url"
})
```

```python
print(new_df.columns)
```

```python
# Keep only the first occurrence of duplicate columns
new_df = new_df.loc[:, ~new_df.columns.duplicated()]

# Optional: confirm
print(new_df.columns)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# -----------------------------
# Step 1: Create 'tags' column
# -----------------------------
new_df["tags"] = (
    new_df["genres"].fillna("") + " " +
    new_df["themes"].fillna("") + " " +
    new_df["synopsis"].fillna("")
).str.lower()

# -----------------------------
# Step 2: TF-IDF vectorization
# -----------------------------
tfidf = TfidfVectorizer(max_features=5000, stop_words="english")
tfidf_matrix = tfidf.fit_transform(new_df["tags"])
```

```python
# -----------------------------
# Step 3: Cosine similarity
# -----------------------------
cosine_sim = cosine_similarity(tfidf_matrix)

# -----------------------------
# Step 4: Map titles to indices
# -----------------------------
indices = pd.Series(new_df.index, index=new_df["title_final"]).drop_duplicates()
```

```python
# -----------------------------
# Step 5: Recommendation function
# -----------------------------
def recommend(title, top_n=10):
    if title not in indices:
        return f"Anime '{title}' not found in dataset."

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]  # exclude the anime itself

    anime_indices = [i[0] for i in sim_scores]

    result = new_df.iloc[anime_indices][[
        "anime_id", "title_final", "score", "anime_url", "image_url"
    ]].copy()

    result["similarity_score"] = [round(i[1], 3) for i in sim_scores]

    result = result.sort_values(by="score", ascending=False).reset_index(drop=True)

    return result.reset_index(drop=True)
```
